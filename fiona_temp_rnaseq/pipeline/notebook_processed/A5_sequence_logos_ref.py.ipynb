```python
import os
import itertools as it
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, Normalize
import seaborn as sns

from seqlogo_utils import (
    get_donor_acceptor_seqs_from_df,
    plot_donor_acceptor_logos,
    plot_u5_u6_heatmap,
    calculate_switch_nt,
)

%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 1
style['ytick.major.size'] = 1
sns.set(font_scale=1.2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
OUTPUT_PATH = snakemake.output.splice_plots
if not os.path.exists(OUTPUT_PATH):
    os.makedirs(OUTPUT_PATH)
```

```python
def convert_coords(coords):
    if not coords:
        return None
    else:
        return np.fromstring(coords.strip('[] '), sep=' ', dtype=int)

psi_fit = pd.read_csv(
    snakemake.input.psi_fit,
    converters={'chrom': str,
                'alt1': convert_coords,
                'alt2': convert_coords}
)
psi_fit_sig = psi_fit.query(
    'event_class == "A5" & (geno_fdr < 0.05 | gxt_fdr < 0.05)'
)
psi_fit_sig.head()
```

```python
a5_alt1_donors, a5_alt1_acceptors, a5_alt2_donors, a5_alt1_acceptors = get_donor_acceptor_seqs_from_df(
    psi_fit_sig, snakemake.input.genome_fasta_fn, winsize=6
)
fig = plt.figure(figsize=(15, 7))
alt1_logo_ax = plt.subplot2grid((3, 2), (0, 0))
alt1_hmap_ax = plt.subplot2grid((3, 2), (1, 0), rowspan=2)
alt2_logo_ax = plt.subplot2grid((3, 2), (0, 1))
alt2_hmap_ax = plt.subplot2grid((3, 2), (1, 1), rowspan=2)

plot_donor_acceptor_logos(
    a5_alt1_donors,
    seq_type='donor',
    title='fio1-3 sensitive 5\'SS',
    ax=alt1_logo_ax
)
_, a5_alt1_u5_grp, a5_alt1_u6_grp = plot_u5_u6_heatmap(a5_alt1_donors, ax=alt1_hmap_ax)

plot_donor_acceptor_logos(
    a5_alt2_donors,
    seq_type='donor',
    title='Alt. 5\'SS increased in fio1-3',
    ax=alt2_logo_ax
)
_, a5_alt2_u5_grp, a5_alt2_u6_grp = plot_u5_u6_heatmap(a5_alt2_donors, vmax=669, ax=alt2_hmap_ax)
plt.savefig(os.path.join(OUTPUT_PATH, 'a5_fio1_logo.svg'))
plt.show()
```

```python
def plot_paired_base_effect_size(alt1_seqs, alt2_seqs, pos, dpsi,
                                 order, pal, iupac_filt,
                                 ax=None):
    # convert 5'SS pos to index
    i = len(alt1_seqs[0]) // 2 + pos - 1
    alt1_base = [iupac_filt[seq[i]] for seq in alt1_seqs]
    alt2_base = [iupac_filt[seq[i]] for seq in alt2_seqs]
    base_grp = [f'{a1}/{a2}' for a1, a2 in zip(alt1_base, alt2_base)]

    mirrored_order = ['/'.join(o) for o in (it.product(order, order[::-1]))]
    if ax is None:
        fig, ax = plt.subplots(figsize=(len(mirrored_order) / 1.5, 4))
    sns.boxplot(
        x=base_grp,
        y=dpsi,
        showfliers=False,
        order=mirrored_order,
        palette=[pal[o[0]] for o in mirrored_order],
        ax=ax
    )
    ax.set_xlabel(f'{pos:+} base of fio1-3 sensitive/\nalternative 5\'SS')
    ax.set_ylabel('Effect size (absolute Î”PSI)')
    return ax


fig = plt.figure(figsize=(11, 5))
p4_ax = plt.subplot2grid((1, 10), (0, 0), colspan=6)
p5_ax = plt.subplot2grid((1, 10), (0, 6), colspan=4, sharey=p4_ax)

plot_paired_base_effect_size(
    a5_alt1_donors, a5_alt2_donors, 4,
    psi_fit_sig.dpsi,
    order='ASU',
    pal={'A': pal[2], 'S': '#888888', 'U': pal[1]},
    iupac_filt={'A': 'A', 'U': 'U', 'G': 'S', 'C': 'S'},
    ax=p4_ax
)

plot_paired_base_effect_size(
    a5_alt1_donors, a5_alt2_donors, 5,
    psi_fit_sig.dpsi,
    order='GH',
    pal={'H': '#888888', 'G': pal[3]},
    iupac_filt={'A': 'H', 'U': 'H', 'G': 'G', 'C': 'H'},
    ax=p5_ax
)
p4_ax.set_ylim(0, 0.5)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_PATH, 'a5_u6_base_effect_size.svg'))
plt.show()
```

```python
order = ['AG', 'BG', 'AH', 'BH']
fig, ax = plt.subplots(figsize=(4, 4))
sns.boxplot(
    x=a5_alt1_u5_grp,
    y=psi_fit_sig.dpsi,
    showfliers=False,
    order=order,
    color=pal[0]
)
ax.set_ylim(0, 0.5)
ax.set_xlabel('U5 class of fio1-3 sensitive 5\'SS')
ax.set_ylabel('Effect size (absolute Î”PSI)')
plt.savefig(os.path.join(OUTPUT_PATH, 'a5_u5_class_effect_size.svg'))
plt.show()
```

```python
# +4 position A vs not A contigency table
a4 = np.array([g[1] for g in a5_alt1_u6_grp + a5_alt2_u6_grp])
y = np.repeat([1, 0], [len(a5_alt1_u6_grp), len(a5_alt2_u6_grp)])

fig, ax = plt.subplots(figsize=(3, 3))
sns.heatmap(
    pd.crosstab(y, a4),
    ax=ax,
    cmap='Blues',
    square=True,
    cbar=False,
    annot=True,
    fmt='d'
)
ax.set_xlabel('Base at +4 position')
ax.set_yticklabels(['Increase', 'Decrease'], va='center')
ax.set_ylabel('Î”PSI in fio1-3')
plt.savefig(os.path.join(OUTPUT_PATH, 'a5_u6_contigency_table_p4.svg'))
plt.show()
```

```python
# full contingency table for +4 position
a4_alt1 = np.array([g[9] for g in a5_alt1_donors])
a4_alt2 = np.array([g[9] for g in a5_alt2_donors])

fig, axes = plt.subplots(figsize=(10, 5), ncols=2)
sns.heatmap(
    pd.crosstab(a4_alt2, a4_alt1),
    ax=axes[0],
    cmap='Blues',
    square=True,
    cbar=False,
    annot=True,
    fmt='d'
)
axes[0].set_xlabel('Base at +4 position in reduced site')
axes[0].set_ylabel('Base at +4 position in increased site')

# -1/+5 contigency table for 5'SS pairs with equivalent +4 positions

mask = [a1 == a2 for a1, a2 in zip(a4_alt1, a4_alt2)]
masked_alt1_m1p5 = [f'{u1[1]}/{u2[2]}' for u1, u2, m
                    in zip(a5_alt1_u5_grp, a5_alt1_u6_grp, mask) if m]
masked_alt2_m1p5 = [f'{u1[1]}/{u2[2]}' for u1, u2, m
                    in zip(a5_alt2_u5_grp, a5_alt2_u6_grp, mask) if m]


contingency_table = pd.crosstab(
        pd.Series(masked_alt2_m1p5, name='Increased 5\'SS -1/+5 base'),
        pd.Series(masked_alt1_m1p5, name='Reduced 5\'SS -1/+5 base'),
)
contingency_table = contingency_table.loc[
    ('H/G', 'H/H', 'G/G', 'G/H'),
    ('H/G', 'G/G', 'H/H', 'G/H'),
]

sns.heatmap(
    contingency_table,
    ax=axes[1],
    cmap='Blues',
    square=True,
    cbar=False,
    annot=True,
    fmt='d'
)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_PATH, 'a5_u6_contigency_table_full.svg'))
plt.show()
```
