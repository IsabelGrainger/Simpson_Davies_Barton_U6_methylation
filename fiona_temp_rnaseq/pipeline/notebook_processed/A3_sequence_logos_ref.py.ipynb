```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, Normalize
import seaborn as sns

from seqlogo_utils import (
    get_donor_acceptor_seqs_from_df,
    plot_donor_acceptor_logos,
    plot_u5_u6_heatmap,
    calculate_switch_nt,
    A3_CARTOON
)

%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 1
style['ytick.major.size'] = 1
sns.set(font_scale=1.2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
sns.palplot(pal)
plt.show()
```

```python
OUTPUT_PATH = snakemake.output.splice_plots
if not os.path.exists(OUTPUT_PATH):
    os.makedirs(OUTPUT_PATH)
```

```python
def convert_coords(coords):
    if not coords:
        return None
    else:
        return np.fromstring(coords.strip('[] '), sep=' ', dtype=int)

psi_fit = pd.read_csv(
    snakemake.input.psi_fit,
    converters={'chrom': str,
                'alt1': convert_coords,
                'alt2': convert_coords}
)
psi_fit_sig = psi_fit.query(
    'event_class == "A3" & (geno_fdr < 0.05 | gxt_fdr < 0.05)'
)
psi_fit_sig.head()
```

```python
a3_alt1_donors, a3_alt1_acceptors, _, a3_alt2_acceptors = get_donor_acceptor_seqs_from_df(
    psi_fit_sig.query('event_class == "A3"'),
    snakemake.input.genome_fasta_fn, winsize=6,
)

fig = plt.figure(figsize=(15, 5))
donor_logo_ax = plt.subplot2grid((4, 2), (1, 0), rowspan=2)
alt1_logo_ax = plt.subplot2grid((4, 2), (0, 1), rowspan=2)
alt2_logo_ax = plt.subplot2grid((4, 2), (2, 1), rowspan=2)

plot_donor_acceptor_logos(
    a3_alt1_donors,
    seq_type='donor',
    title='5\'SS for all fio1-3 sensitive alt. 3\'SS',
    ax=donor_logo_ax
)
plot_donor_acceptor_logos(
    a3_alt1_acceptors,
    seq_type='acceptor',
    title='fio1-3 sensitive 3\'SS',
    ax=alt1_logo_ax
)
plot_donor_acceptor_logos(
    a3_alt2_acceptors,
    seq_type='acceptor',
    title='Alt. 3\'SS increased in fio1-3',
    ax=alt2_logo_ax
)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_PATH, 'a3_fio1_all_logo_heatmap.svg'))
plt.show()
```

### Alternative 3'SS switching upstream and downstream

```python
a3_switch = psi_fit_sig.query('event_class == "A3"').apply(calculate_switch_nt, which='a3', axis=1)

fig, ax = plt.subplots(figsize=(10, 4))
ax.hist(a3_switch, bins=np.arange(-60, 61), align='left')
ax.set_ylabel('Alt 3\'SS pairs')
ax.set_xlabel('Distance between alternative 3\'SS pairs with reduced/increased usage in fio1-3')
plt.savefig(os.path.join(OUTPUT_PATH, 'a3_switch_size_histogram.svg'))
plt.show()
```

```python
a3u_alt1_donors, a3u_alt1_acceptors, _, a3u_alt2_acceptors = get_donor_acceptor_seqs_from_df(
    psi_fit_sig.query('event_class == "A3"')[a3_switch < 0],
    snakemake.input.genome_fasta_fn, winsize=6,
)

a3d_alt1_donors, a3d_alt1_acceptors, _, a3d_alt2_acceptors = get_donor_acceptor_seqs_from_df(
    psi_fit_sig.query('event_class == "A3"')[a3_switch > 0],
    snakemake.input.genome_fasta_fn, winsize=6,
)

fig = plt.figure(figsize=(20, 7))
a3u_donor_logo_ax = plt.subplot2grid((3, 3), (0, 0))
a3u_ups_logo_ax = plt.subplot2grid((3, 3), (0, 1))
a3u_dwn_logo_ax = plt.subplot2grid((3, 3), (0, 2))
cartoon_ax = plt.subplot2grid((3, 3), (1, 0), colspan=3)
a3d_donor_logo_ax = plt.subplot2grid((3, 3), (2, 0))
a3d_ups_logo_ax = plt.subplot2grid((3, 3), (2, 1))
a3d_dwn_logo_ax = plt.subplot2grid((3, 3), (2, 2))


# plot_cartoon in centre
cartoon_ax.set_axis_off()
cartoon_ax.imshow(A3_CARTOON)

plot_donor_acceptor_logos(
    a3u_alt1_donors,
    seq_type='donor',
    ax=a3u_donor_logo_ax
)
plot_donor_acceptor_logos(
    a3u_alt2_acceptors,
    seq_type='acceptor',
    ax=a3u_ups_logo_ax,
)
plot_donor_acceptor_logos(
    a3u_alt1_acceptors,
    seq_type='acceptor',
    ax=a3u_dwn_logo_ax,
)
a3u_donor_logo_ax.text(
    .5, 0.93,
    f'3\'SS pairs with increased upstream use in fio1-3 (n={len(a3u_alt1_donors)})',
    transform=fig.transFigure,
    horizontalalignment='center'
)


plot_donor_acceptor_logos(
    a3d_alt1_donors,
    seq_type='donor',
    ax=a3d_donor_logo_ax
)
plot_donor_acceptor_logos(
    a3d_alt1_acceptors,
    seq_type='acceptor',
    ax=a3d_ups_logo_ax,
)
plot_donor_acceptor_logos(
    a3d_alt2_acceptors,
    seq_type='acceptor',
    ax=a3d_dwn_logo_ax,
)
a3d_donor_logo_ax.text(
    .5, 0.03,
    f'3\'SS pairs with increased downstream use in fio1-3 (n={len(a3d_alt1_donors)})',
    transform=fig.transFigure,
    horizontalalignment='center'
)
plt.savefig(os.path.join(OUTPUT_PATH, 'a3_fio1_logo.svg'))
plt.show()
```
