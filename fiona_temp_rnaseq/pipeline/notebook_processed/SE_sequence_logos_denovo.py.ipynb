```python
OUTPUT_PATH = snakemake.output.splice_plots
if not os.path.exists(OUTPUT_PATH):
    os.makedirs(OUTPUT_PATH)
```

```python
def display_formatted_markdown(md, **kwargs):
    md_f = md.format(**kwargs)
    display_markdown(Markdown(md_f))

MD_TEXT = open('../rules/notebook_templates/md_text/SE.md').readlines()

ref_type = {'ref': 'reference', 'denovo': '*de novo*'}

display_formatted_markdown(
    '# Skipped exon analysis using {ref_type} transcriptome assembly',
    ref_type=ref_type[snakemake.wildcards.ref_type]
)
```

```python
def convert_coords(coords):
    if not coords:
        return None
    else:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            return np.fromstring(coords.strip('[] '), sep=' ', dtype=int)

psi_fit = pd.read_csv(
    snakemake.input.psi_fit,
    converters={'chrom': str,
                'alt1': convert_coords,
                'alt2': convert_coords}
)
psi_fit_sig = psi_fit.query(
    'event_class == "SE" & (geno_fdr < 0.05 | gxt_fdr < 0.05)'
)
```

```python
display_formatted_markdown(
    MD_TEXT[0],
    num_se=len(psi_fit_sig),
    perc_inc=np.mean(psi_fit_sig.dpsi > 0) * 100,
    perc_skip=np.mean(psi_fit_sig.dpsi < 0) * 100,
)

se_pos, se_seqs = get_donor_acceptor_seqs_from_df(
    psi_fit_sig.query('dpsi < 0'),
    snakemake.input.genome_fasta_fn, winsize=6,
)

ie_pos, ie_seqs = get_donor_acceptor_seqs_from_df(
    psi_fit_sig.query('dpsi > 0'),
    snakemake.input.genome_fasta_fn, winsize=6,
)

se_donors, use_donors, ie_donors, uie_donors = (
    list(se_seqs['alt1']['donor'].values()),
    list(se_seqs['alt2']['donor'].values()),
    list(ie_seqs['alt1']['donor'].values()),
    list(ie_seqs['alt2']['donor'].values()),
)

display_formatted_markdown(
    MD_TEXT[1],
    se_perc_a4=perc_seq_pos(se_donors, 4, 'A')[1],
)

fig = plt.figure(figsize=(15, 7))
use_logo_ax = plt.subplot2grid((3, 2), (0, 0))
se_logo_ax = plt.subplot2grid((3, 2), (0, 1))
cartoon_ax = plt.subplot2grid((3, 2), (1, 0), colspan=2)
uie_logo_ax = plt.subplot2grid((3, 2), (2, 0))
ie_logo_ax = plt.subplot2grid((3, 2), (2, 1))

# plot_cartoon in centre
cartoon_ax.set_axis_off()
cartoon_ax.imshow(SE_CARTOON)

plot_donor_acceptor_logos(
    use_donors,
    seq_type='donor',
    ax=use_logo_ax,
    title='',
)

plot_donor_acceptor_logos(
    se_donors,
    seq_type='donor',
    ax=se_logo_ax,
    title='',
)
se_logo_ax.text(
    .5, 0.93,
    f'Exons with increased skipping in fio1-3',
    transform=fig.transFigure,
    horizontalalignment='center'
)

plot_donor_acceptor_logos(
    uie_donors,
    seq_type='donor',
    ax=uie_logo_ax,
    title='',
)


plot_donor_acceptor_logos(
    ie_donors,
    seq_type='donor',
    ax=ie_logo_ax,
    title='',
)
ie_logo_ax.text(
    .5, 0.03,
    f'Exons with increased inclusion in fio1-3',
    transform=fig.transFigure,
    horizontalalignment='center'
)

plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_PATH, 'se_fio1_logo.svg'))
plt.show()

display_formatted_markdown(
    MD_TEXT[2],
    ie_perc_u4=perc_seq_pos(ie_donors, 4, 'U')[1],
    ie_perc_a4=perc_seq_pos(ie_donors, 4, 'A')[1],
    uie_perc_a4=perc_seq_pos(uie_donors, 4, 'A')[1],
)
```
