```python
OUTPUT_PATH = snakemake.output.splice_plots
if not os.path.exists(OUTPUT_PATH):
    os.makedirs(OUTPUT_PATH)
```

```python
def display_formatted_markdown(md, **kwargs):
    md_f = md.format(**kwargs)
    display_markdown(Markdown(md_f))

MD_TEXT = open('../rules/notebook_templates/md_text/A3.md').readlines()

ref_type = {'ref': 'reference', 'denovo': '*de novo*'}

display_formatted_markdown(
    '# Alternative 3\'SS analysis using {ref_type} transcriptome assembly',
    ref_type=ref_type[snakemake.wildcards.ref_type]
)
```

```python
def convert_coords(coords):
    if not coords:
        return None
    else:
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")
            return np.fromstring(coords.strip('[] '), sep=' ', dtype=int)

psi_fit = pd.read_csv(
    snakemake.input.psi_fit,
    converters={'chrom': str,
                'alt1': convert_coords,
                'alt2': convert_coords}
)
psi_fit_sig = psi_fit.query(
    'event_class == "A3" & (geno_fdr < 0.05 | gxt_fdr < 0.05)'
)
```

```python
a3_pos, a3_seqs = get_donor_acceptor_seqs_from_df(
    psi_fit_sig, snakemake.input.genome_fasta_fn, winsize=6,
)

a3_donors = list(a3_seqs['alt1']['donor'].values())
a3_alt1_acceptors = list(a3_seqs['alt1']['acceptor'].values())
a3_alt2_acceptors = list(a3_seqs['alt2']['acceptor'].values())


display_formatted_markdown(
    MD_TEXT[0],
    num_a3=len(psi_fit_sig),
)

fig = plt.figure(figsize=(15, 5))
donor_logo_ax = plt.subplot2grid((4, 2), (1, 0), rowspan=2)
alt1_logo_ax = plt.subplot2grid((4, 2), (0, 1), rowspan=2)
alt2_logo_ax = plt.subplot2grid((4, 2), (2, 1), rowspan=2)

plot_donor_acceptor_logos(
    a3_donors,
    seq_type='donor',
    title='5\'SS for all fio1-3 sensitive alt. 3\'SS',
    ax=donor_logo_ax
)
plot_donor_acceptor_logos(
    a3_alt1_acceptors,
    seq_type='acceptor',
    title='fio1-3 sensitive 3\'SS',
    ax=alt1_logo_ax
)
plot_donor_acceptor_logos(
    a3_alt2_acceptors,
    seq_type='acceptor',
    title='Alt. 3\'SS increased in fio1-3',
    ax=alt2_logo_ax
)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_PATH, 'a3_fio1_all_logo_heatmap.svg'))
plt.show()

a3_switch = psi_fit_sig.query('event_class == "A3"').apply(calculate_switch_nt, which='a3', axis=1)

display_formatted_markdown(
    MD_TEXT[1],
    a3_5nt=np.mean(a3_switch.abs() <= 5) * 100,
    a3_3nt_exact=np.mean(a3_switch.abs() == 3) * 100,
)

fig, ax = plt.subplots(figsize=(10, 4))
ax.hist(a3_switch, bins=np.arange(-60, 61), align='left')
ax.set_ylabel('Alt 3\'SS pairs')
ax.set_xlabel('Distance between alternative 3\'SS pairs with reduced/increased usage in fio1-3')
plt.savefig(os.path.join(OUTPUT_PATH, 'a3_switch_size_histogram.svg'))
plt.show()
```

```python
a3u_pos, a3u_seqs = get_donor_acceptor_seqs_from_df(
    psi_fit_sig.query('event_class == "A3"')[a3_switch < 0],
    snakemake.input.genome_fasta_fn, winsize=6,
)

a3u_donors, a3u_alt1_acceptors, a3u_alt2_acceptors = (
    list(a3u_seqs['alt1']['donor'].values()),
    list(a3u_seqs['alt1']['acceptor'].values()),
    list(a3u_seqs['alt2']['acceptor'].values()),
)

a3d_pos, a3d_seqs = get_donor_acceptor_seqs_from_df(
    psi_fit_sig.query('event_class == "A3"')[a3_switch > 0],
    snakemake.input.genome_fasta_fn, winsize=6,
)

a3d_donors, a3d_alt1_acceptors, a3d_alt2_acceptors = (
    list(a3d_seqs['alt1']['donor'].values()),
    list(a3d_seqs['alt1']['acceptor'].values()),
    list(a3d_seqs['alt2']['acceptor'].values()),
)

display_formatted_markdown(
    MD_TEXT[2],
    a3u_perc_a4=perc_seq_pos(a3u_donors, 4, 'A')[1],
    a3u_perc_d3c=perc_seq_pos(a3u_alt1_acceptors, -2, 'C')[1], # -2 because perc_seq_pos was written for donors not acceptors so there is an off by one error
    a3u_perc_u3c=perc_seq_pos(a3u_alt2_acceptors, -2, 'C')[1],
)
display_formatted_markdown(
    MD_TEXT[3],
    a3d_perc_a4=perc_seq_pos(a3d_donors, 4, 'A')[1],
    a3d_perc_u4=perc_seq_pos(a3d_donors, 4, 'U')[1],
)

fig = plt.figure(figsize=(20, 7))
a3u_donor_logo_ax = plt.subplot2grid((3, 3), (0, 0))
a3u_ups_logo_ax = plt.subplot2grid((3, 3), (0, 1))
a3u_dwn_logo_ax = plt.subplot2grid((3, 3), (0, 2))
cartoon_ax = plt.subplot2grid((3, 3), (1, 0), colspan=3)
a3d_donor_logo_ax = plt.subplot2grid((3, 3), (2, 0))
a3d_ups_logo_ax = plt.subplot2grid((3, 3), (2, 1))
a3d_dwn_logo_ax = plt.subplot2grid((3, 3), (2, 2))


# plot_cartoon in centre
cartoon_ax.set_axis_off()
cartoon_ax.imshow(A3_CARTOON)

plot_donor_acceptor_logos(
    a3u_donors,
    seq_type='donor',
    ax=a3u_donor_logo_ax,
    title=''
)
plot_donor_acceptor_logos(
    a3u_alt2_acceptors,
    seq_type='acceptor',
    ax=a3u_ups_logo_ax,
    title=''
)
plot_donor_acceptor_logos(
    a3u_alt1_acceptors,
    seq_type='acceptor',
    ax=a3u_dwn_logo_ax,
    title=''
)
a3u_donor_logo_ax.text(
    .5, 0.93,
    f'Alt. 3\'SSs with increased upstream use in fio1-3',
    transform=fig.transFigure,
    horizontalalignment='center'
)


plot_donor_acceptor_logos(
    a3d_donors,
    seq_type='donor',
    ax=a3d_donor_logo_ax,
    title=''
)
plot_donor_acceptor_logos(
    a3d_alt1_acceptors,
    seq_type='acceptor',
    ax=a3d_ups_logo_ax,
    title=''
)
plot_donor_acceptor_logos(
    a3d_alt2_acceptors,
    seq_type='acceptor',
    ax=a3d_dwn_logo_ax,
    title=''
)
a3d_donor_logo_ax.text(
    .5, 0.03,
    f'Alt. 3\'SSs with increased downstream use in fio1-3',
    transform=fig.transFigure,
    horizontalalignment='center'
)
plt.savefig(os.path.join(OUTPUT_PATH, 'a3_fio1_logo.svg'))
plt.show()
```

```python
psi_a3 = pd.read_csv(
    snakemake.input.psi,
    header=[0, 1],
    skiprows=[2,],
    index_col=0,
    converters={2: str,
                3: convert_coords,
                4: convert_coords}
).droplevel(0, axis=1).query('event_class == "A3"')
psi_a3['a3_switch'] = psi_a3.apply(calculate_switch_nt, which='a3', axis=1)
psi_a3_6nt = psi_a3.query('abs(a3_switch) <= 6')
a3_6nt_seqs = get_pairwise_seqs(*get_donor_acceptor_seqs_from_df(
    psi_a3_6nt, snakemake.input.genome_fasta_fn, winsize=6,
), which='acceptor')
a3_alt2_psi = psi_a3_6nt.loc[:, psi_a3_6nt.columns.str.startswith('col0_20c')].mean(1)
ups_seq = []
dwn_seq = []
dwn_psi = []
for alt1, alt2, alt2_psi, switch in zip(*a3_6nt_seqs, a3_alt2_psi, psi_a3_6nt.a3_switch):
    if switch < 0:
        dwn_seq.append(alt1)
        dwn_psi.append(1 - alt2_psi)
        ups_seq.append(alt2)
    else:
        dwn_seq.append(alt2)
        dwn_psi.append(alt2_psi)
        ups_seq.append(alt1)
a3_up_dwn_psi = pd.DataFrame.from_dict(
    {'upstream_seq': ups_seq, 'downstream_seq': dwn_seq, 'downstream_psi': dwn_psi}
)
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(
    hue=a3_up_dwn_psi.upstream_seq.str[3],
    x=a3_up_dwn_psi.downstream_seq.str[3],
    y=a3_up_dwn_psi.downstream_psi,
    order=['C', 'U', 'A', 'G'],
    hue_order=['C', 'U', 'A', 'G'],
    showfliers=False,
    ax=ax
)
ax.legend(title='Upstream 3\'SS\n-3 position', ncol=2)
ax.set_xlabel('Downstream 3\'SS -3 position')
ax.set_ylabel('Downstream PSI')
plt.savefig(os.path.join(OUTPUT_PATH, 'a3_downstream_vs_upstream_psi.svg'))
plt.show()
```
