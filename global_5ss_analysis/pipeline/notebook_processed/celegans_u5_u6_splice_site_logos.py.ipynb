```python
RC = str.maketrans('ACGT', 'TGCA')


def rev_comp(seq):
    return seq.translate(RC)[::-1]


def get_gtf_attribute(gtf_record, attribute):
    try:
        attr = re.search(f'{attribute} "(.+?)";', gtf_record[8]).group(1)
    except AttributeError:
        raise ValueError(
            f'Could not parse attribute {attribute} '
            f'from GTF with feature type {record[2]}'
        )
    return attr


def gtf_parser(gtf_fn):
    gtf_records = {}
    with open(gtf_fn) as gtf:
        for i, record in enumerate(gtf):
            if record.startswith('#') or record.startswith('track'):
                continue
            record = record.split('\t')
            chrom, _, feat_type, start, end, _, strand = record[:7]
            start = int(start) - 1
            end = int(end)
            if feat_type == 'exon':
                transcript_id = get_gtf_attribute(record, 'transcript_id')
                gene_id = get_gtf_attribute(record, 'gene_id')
                idx = (chrom, gene_id, transcript_id, strand)
                if idx not in gtf_records:
                    gtf_records[idx] = []
                gtf_records[idx].append([start, end])
    for (chrom, gene_id, transcript_id, strand), invs in gtf_records.items():
        invs.sort()
        exons = np.array(invs)
        yield chrom, gene_id, transcript_id, strand, exons


def get_all_intron_donor_acceptor_seqs(gtf_fn, fasta_fn, winsize=6):
    donor_acceptor_seqs = {}
    with pysam.FastaFile(fasta_fn) as fasta:
        for chrom, gene_id, transcript_id, strand, exons in gtf_parser(gtf_fn):
            donors = exons[:-1, 1]
            acceptors = exons[1:, 0]
            if strand == '-':
                donors, acceptors = acceptors, donors
            for donor, acceptor in zip(donors, acceptors):
                dstr = f'{chrom}:{donor}{strand}'
                astr = f'{chrom}:{acceptor}{strand}'
                ds = fasta.fetch(chrom, donor - winsize, donor + winsize)
                as_ = fasta.fetch(chrom, acceptor - winsize, acceptor + winsize)
                if strand == '-':
                    ds = rev_comp(ds)
                    as_ = rev_comp(as_)
                donor_acceptor_seqs[(dstr, astr)] = [
                    ds.upper().replace('T', 'U'),
                    as_.upper().replace('T', 'U'),
                    abs(donor - acceptor)
                ]
    donor_seqs, acceptor_seqs, intron_lengths = zip(*donor_acceptor_seqs.values())
    return donor_seqs, acceptor_seqs, intron_lengths
```

```python
RC = str.maketrans('ACGT', 'TGCA')


def rev_comp(seq):
    return seq.translate(RC)[::-1]


IUPAC = {
    'A': 'A', 'C': 'C', 'G': 'G', 'U': 'U',
    'S': 'GC', 'W': 'AU', 'R': 'AG', 'Y': 'CU',
    'K': 'GU', 'M': 'AC', 'N': 'ACGU',
}

IUPAC_INV = {
    'A': 'B', 'C': 'D', 'G': 'H', 'U': 'V',
    'S': 'W', 'W': 'S', 'R': 'Y', 'Y': 'R',
    'K': 'M', 'M': 'K',
    'N': 'N'
}


def iupac_classify(seq, consensus):
    clss = []
    for o, e in zip(seq, consensus):
        if o in IUPAC[e]:
            clss.append(e)
        
        else:
            clss.append(IUPAC_INV[e])
    return ''.join(clss)


def u5_classify(donor_seq):
    assert not len(donor_seq) % 2
    ws = len(donor_seq) // 2
    return iupac_classify(donor_seq[ws - 2: ws], 'AG')


def u6_classify(donor_seq):
    assert not len(donor_seq) % 2
    ws = len(donor_seq) // 2
    return iupac_classify(donor_seq[ws + 2: ws + 5], 'RAG')


def acceptor_classify(acceptor_seq):
    assert not len(acceptor_seq) % 2
    ws = len(acceptor_seq) // 2
    return iupac_classify(acceptor_seq[ws - 5: ws - 2], 'UKC')


def edit_distance(seq1, seq2):
    ed = 0
    for i, j in zip(seq1, seq2):
        if i != j:
            ed += 1
    return ed


IUPAC_OHE = {
    'A': [1, 0, 0, 0],
    'C': [0, 1, 0, 0],
    'G': [0, 0, 1, 0],
    'U': [0, 0, 0, 1],
    'R': [0.5, 0, 0.5, 0],
    'Y': [0, 0.5, 0, 0.5],
    'S': [0, 0.5, 0.5, 0],
    'W': [0.5, 0, 0, 0.5],
    'B': [0, 1/3, 1/3, 1/3],
    'H': [1/3, 1/3, 0, 1/3],
    'N': [0.25, 0.25, 0.25, 0.25]
}


def ohe_seq_iupac(seq):
    seq_ohe = []
    for n in seq:
        seq_ohe.append(IUPAC_OHE[n])
    return np.array(seq_ohe)


def pwm_score(seq, pwm):
    pwm = np.array(pwm)
    seq_ohe = ohe_seq_iupac(seq)
    return (seq_ohe * pwm).ravel().sum()
```

```python
def plot_donor_acceptor_logos(seqs, seq_type='donor', title=None, ax=None):
    w = len(seqs[0]) // 2
    if seq_type == 'donor':
        trim_method = lambda seqs: [s[w - 2: w + 5] for s in seqs]
        xticks = np.arange(7) + 0.5
        xticklabels = ['−2', '−1', '+1', '+2', '+3', '+4', '+5']
    elif seq_type == 'acceptor':
        trim_method = lambda seqs: [s[w - 5: w + 2] for s in seqs]
        xticks = np.arange(7) + 0.5
        xticklabels = ['−5', '−4', '−3', '−2', '−1', '+1', '+2']
    else:
        trim_method = lambda seqs: seqs
        xticks = []
        xticklabels = []
    ax = logo.draw_logo(
        trim_method(seqs),
        alphabet='rna',
        y_format='probability',
        ax=ax,
    )
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xticks(xticks)
    ax.tick_params(bottom=False)
    ax.set_xticklabels(xticklabels)
    if title is not None:
        ax.set_title(f'{title} (n={len(seqs)})')
    return ax


def all_u5_u6_classifications(fasta_fn, gtf_fn):
    print('loading donor/acceptor seqs')
    donor_seqs, acceptor_seqs, intron_lengths = get_all_intron_donor_acceptor_seqs(
        gtf_fn,
        fasta_fn
    )
    ax = plot_donor_acceptor_logos(donor_seqs)
    plt.show()
    ax = plot_donor_acceptor_logos(acceptor_seqs, seq_type='acceptor')
    plt.show()
    donor_pwm = np.log(calculate_normalised_counts(donor_seqs, ALPHABETS['rna']))
    pwms = {'U5': donor_pwm[4:6], 'U6': donor_pwm[8:11]}

    u5_u6_classifications = []

    for seq, ilen in zip(donor_seqs, intron_lengths):
        u5_u6_classifications.append(
            [seq, 
             u5_classify(seq), pwm_score(seq[4:6], pwms['U5']),
             u6_classify(seq), pwm_score(seq[8:11], pwms['U6']),
             ilen]
        )

    u5_u6_classifications = pd.DataFrame(
        u5_u6_classifications,
        columns=['seq', 'u5_class', 'u5_pwm_score',
                 'u6_class', 'u6_pwm_score', 'intron_length']
    )
    print('U5 vs U6 score correlation')
    print(stats.spearmanr(
        u5_u6_classifications.u5_pwm_score,
        u5_u6_classifications.u6_pwm_score,
    ))
    return donor_seqs, acceptor_seqs, u5_u6_classifications, pwms
```

```python
def plot_ecdf(data, ax=None, color=None, **plot_kwargs):
    x = np.unique(data)
    ecdf = np.cumsum([(data == n).sum() for n in x]) / len(data)
    if ax is None:
        fig, ax = plt.subplots()
    ax.step(x, ecdf, where='post', color=color, **plot_kwargs)
    return ax


def plot_u6_u5_ecdfs(classifications):
    fig, axes = plt.subplots(figsize=(8.5, 4), ncols=2, sharey=True)
    plot_ecdf(
        classifications.query('u6_class == "RAG"').u5_pwm_score,
        ax=axes[0],
        lw=3,
        alpha=0.8,
        label='//GURAG',
        color=pal[0],
    )
    plot_ecdf(
        classifications.query('u6_class != "RAG"').u5_pwm_score,
        ax=axes[0],
        lw=3,
        alpha=0.8,
        label='Other',
        color=pal[1],
    )
    axes[0].legend(title='U6 class', loc=2)
    axes[0].set_xlabel('U5 PSSM log-likelihood')

    for i, u5_cls in enumerate(['AG', 'BG', 'AH', 'BH']):
        plot_ecdf(
            classifications.query('u5_class == @u5_cls').u6_pwm_score,
            ax=axes[1],
            lw=3,
            alpha=0.8,
            label=u5_cls + '//GU',
            color=pal[i],
        )
    axes[1].legend(title='U5 class', loc=2)
    axes[1].set_xlabel('U6 PSSM log-likelihood')
    return fig, axes
```

```python
def get_annot_alt_seqs(suppa_alt5_gtf_fn, fasta_fn, winsize=6, which='a5', max_dist=np.inf):
    seq_pairs = {}
    other_seq = {}
    with pysam.FastaFile(fasta_fn) as fasta:
        for chrom, _, event_id, strand, invs in gtf_parser(suppa_alt5_gtf_fn):
            event_id, alt_id = re.split('[.:]alt(?:ernative)?', event_id)
            if strand == '+':
                pos = invs[0, 1] if which == 'a5' else invs[1, 0]
                opos = invs[1, 0] if which == 'a5' else invs[0, 1]
            else:
                pos = invs[1, 0] if which == 'a5' else invs[0, 1]
                opos = invs[0, 1] if which == 'a5' else invs[1, 0]
            seq = fasta.fetch(chrom, pos - winsize, pos + winsize).upper()
            oseq = fasta.fetch(chrom, opos - winsize, opos + winsize).upper()
            if strand == '-':
                seq = rev_comp(seq)
                oseq = rev_comp(oseq)
            if event_id not in seq_pairs:
                seq_pairs[event_id] = []
            seq_pairs[event_id].append([seq.replace('T', 'U'), pos, strand])
            other_seq[event_id] = oseq.replace('T', 'U')
    # order by more upstream
    seq_pairs_ordered = []
    other_seq_ordered = []
    for event_id, pairs in seq_pairs.items():
        seq1, alt1_pos, strand = pairs.pop()
        seq2, alt2_pos, _ = pairs.pop()
        if abs(alt1_pos - alt2_pos) > max_dist:
            continue
        if alt1_pos < alt2_pos:
            if strand == '+':
                seq_pairs_ordered.append([seq1, seq2])
            else:
                seq_pairs_ordered.append([seq2, seq1])
        else:
            if strand == '+':
                seq_pairs_ordered.append([seq2, seq1])
            else:
                seq_pairs_ordered.append([seq1, seq2])
        other_seq_ordered.append(other_seq[event_id])
    return seq_pairs_ordered, other_seq_ordered


def alt_donor_pair_scores(a5_gtf, fasta_fn, pwms):
    alt_donor_pairs, alt_donor_pair_acceptor = get_annot_alt_seqs(
        a5_gtf, fasta_fn
    )
    u5_u6_classifications = []

    for seq1, seq2 in alt_donor_pairs:
        u5_u6_classifications.append(
            [seq1, seq2,
             u5_classify(seq1), pwm_score(seq1[4:6], pwms['U5']),
             u5_classify(seq2), pwm_score(seq2[4:6], pwms['U5']),
             u6_classify(seq1), pwm_score(seq1[8:11], pwms['U6']),
             u6_classify(seq2), pwm_score(seq2[8:11], pwms['U6'])]
        )

    u5_u6_classifications = pd.DataFrame(
        u5_u6_classifications,
        columns=['upstream_seq', 'downstream_seq',
                 'upstream_u5_class', 'upstream_u5_pwm_score',
                 'downstream_u5_class', 'downstream_u5_pwm_score',
                 'upstream_u6_class', 'upstream_u6_pwm_score',
                 'downstream_u6_class', 'downstream_u6_pwm_score']
    )
    u5_u6_classifications['u5_log_odds'] = (
        u5_u6_classifications.upstream_u5_pwm_score - u5_u6_classifications.downstream_u5_pwm_score
    )
    u5_u6_classifications['u6_log_odds'] = (
        u5_u6_classifications.upstream_u6_pwm_score - u5_u6_classifications.downstream_u6_pwm_score
    )
    return u5_u6_classifications
```

```python
def alt_donor_pair_scatterplot(a5_gtf, fasta_fn, pwms):
    
    u5_u6_classifications = alt_donor_pair_scores(a5_gtf, fasta_fn, pwms)
    fig = plt.figure(figsize=(12.5, 5))
    scatter_ax = plt.subplot2grid(shape=(12, 13), loc=(0, 0), rowspan=12, colspan=5)
    logo_ax1 = plt.subplot2grid(shape=(12, 13), loc=(0, 6), rowspan=5, colspan=7)
    logo_ax2 = plt.subplot2grid(shape=(12, 13), loc=(7, 6), rowspan=5, colspan=7)
    
    sns.regplot(
        x='u5_log_odds',
        y='u6_log_odds',
        data=u5_u6_classifications,
        ax=scatter_ax,
        scatter_kws=dict(color=pal[-2], alpha=0.25, s=4, marker='x', rasterized=True),
        line_kws=dict(color=pal[0])
    )
    spearmn = stats.spearmanr(u5_u6_classifications.u5_log_odds, u5_u6_classifications.u6_log_odds)
    print(spearmn)
    scatter_ax.annotate(xy=(0.05, 0.05), s=f'Spearmans rho={spearmn[0]:.2f}', xycoords='axes fraction')
    scatter_ax.set_xlabel('U5 PSSM log odds')
    scatter_ax.set_ylabel('U6 PSSM log odds')

    u6_stronger = []
    u6_weaker = []
    for _, ups, dwns, u6d in u5_u6_classifications[['upstream_seq', 'downstream_seq', 'u6_log_odds']].itertuples():
        if u6d < 0:
            u6_stronger.append(dwns)
            u6_weaker.append(ups)
        elif u6d >= 0:
            u6_stronger.append(ups)
            u6_weaker.append(dwns)

    plot_donor_acceptor_logos(u6_stronger, title='Alt 5\'SS with higher U6 PSSM score', ax=logo_ax1)
    plot_donor_acceptor_logos(u6_weaker, title='Alt 5\'SS with lower U6 PSSM score', ax=logo_ax2)

    return fig, [scatter_ax, logo_ax1, logo_ax2]
```

```python
fig, axes = alt_donor_pair_scatterplot(
    snakemake.input.a5_gtf,
    snakemake.input.fasta,
    pwms
)
plt.savefig(snakemake.output.u5_u6_altsplice_scatter)
plt.show()
```
