```python
import sys
import os
from glob import glob
import random
import re
import itertools as it
from collections import Counter
import json
import gzip

import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib_logo import draw_logo
from matplotlib.colors import ListedColormap
import seaborn as sns
from IPython.display import Markdown, display_markdown
import pysam

from gene_track_utils import count_sites_in_genic_features

## Default plotting params

%matplotlib inline
sns.set(font='Arial')
plt.rcParams['svg.fonttype'] = 'none'
style = sns.axes_style('white')
style.update(sns.axes_style('ticks'))
style['xtick.major.size'] = 2
style['ytick.major.size'] = 2
sns.set(font_scale=1.2, style=style)
pal = sns.color_palette(['#0072b2', '#d55e00', '#009e73', '#f0e442', '#cc79a7', '#56b4e9', '#e69f00'])
cmap = ListedColormap(pal.as_hex())
sns.set_palette(pal)
```

```python
def display_formatted_markdown(md, **kwargs):
    md_f = md.format(**kwargs)
    display_markdown(Markdown(md_f))

genotypes = {'col0': 'Col-0', 'fio1': '*fio1-3*', 'fip37': '*fip37-4*'}    

def format_comp(comp):
    if '__not__' in comp:
        comps = comp.split('__not__')
        comp1 = comps[0].split('_vs_')
        comp2 = comps[1].split('_vs_')
        if comp2[1].endswith('__miclip'):
            comp2[1] = comp2[1][:-8]
            adds = ' but are found in miCLIP'
        else:
            adds = ''
        return f'{genotypes[comp1[0]]} vs. {genotypes[comp1[1]]} which are not found in {genotypes[comp2[0]]} vs. {genotypes[comp2[1]]}' + adds
    elif '__and__' in comp:
        comps = comp.split('__and__')
        comp1 = comps[0].split('_vs_')
        comp2 = comps[1].split('_vs_')
        return f'{genotypes[comp1[0]]} vs. {genotypes[comp1[1]]} and also in comparison {genotypes[comp2[0]]} vs. {genotypes[comp2[1]]}'
    else:
        comp = comp.split('_vs_')
        return f'{genotypes[comp[0]]} vs. {genotypes[comp[1]]}'

MD_TEXT = open('../rules/notebook_templates/md_text/logos.md').readlines()

display_formatted_markdown(
    '# Sequence motif and distribution analysis for sites found in comparison {comp}',
    comp=format_comp(snakemake.wildcards.comp)
)
```

```python
RC = str.maketrans('ACGT', 'TGCA')


def rev_comp(seq):
    return seq.translate(RC)[::-1]


motifs = pd.read_csv(
    snakemake.input.motifs,
    sep='\t', header=None,
    names=['chrom', 'start', 'end', 'motif', 'score', 'strand']
)
most_common = motifs.motif.value_counts().index[0]
# try to phase drach motif
if re.match('[ACGT][AGT][AG]A[CT]', most_common):
    offset = 1
else:
    offset = 0

motif_seqs = []

with pysam.FastaFile(snakemake.input.fasta) as fasta:
    for _, record in motifs.iterrows():
        o = offset if record.strand == '+' else -offset
        seq = fasta.fetch(str(record.chrom), record.start + o, record.end + o)
        if record.strand == '-':
            seq = rev_comp(seq)
        motif_seqs.append(seq.replace('T', 'U'))

display_formatted_markdown(
    MD_TEXT[0],
    n_sites=len(pd.read_csv(snakemake.input.sites, sep='\t')),
    comp=format_comp(snakemake.wildcards.comp),
    n_motifs=len(motifs),
)

ax = draw_logo(motif_seqs, alphabet='rna', y_format='probability', fig_height=3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title(f'm6A site motifs (n={len(motif_seqs)})')
plt.savefig(snakemake.output.logo_plot)
plt.show()
```

```python
tmp_filename = os.path.join(os.environ['TMPDIR'], os.path.split(f'{snakemake.input.sites}.gz')[1])
pysam.tabix_compress(snakemake.input.sites, tmp_filename, force=True)
pysam.tabix_index(tmp_filename, force=True, preset='bed')

feat_count = count_sites_in_genic_features(tmp_filename, snakemake.input.gtf)

display_formatted_markdown(
    MD_TEXT[1],
    main_feature=feat_count.groupby('feat_type').mean().idxmax()[0]
)

fig, ax = plt.subplots(figsize=(5, 5))
sns.barplot(
    x='feat_type',
    y='sites_per_kb',
    data=feat_count,
    order=['Upstream', '5\'UTR', 'CDS', 'Introns', '3\'UTR', 'Downstream'],
    color=pal[0],
)
ax.set_xticklabels(['Upstream', '5\'UTR', 'CDS', 'Intron', '3\'UTR', 'Downstream'], rotation=35, ha='right')
ax.set_xlabel('Feature type')
ax.set_ylabel('Average modification sites per kb')
plt.savefig(snakemake.output.distrib_plot)
plt.show()
```
