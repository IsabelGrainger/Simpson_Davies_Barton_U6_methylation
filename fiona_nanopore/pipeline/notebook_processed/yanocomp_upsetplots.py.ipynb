```python
YANOCOMP_COLUMNS = [
    'chrom', 'start', 'end', 'gene_id', 'kmer', 'score', 'strand',
    'log_odds', 'log_odds_lower_ci', 'log_odds_upper_ci',
    'pval', 'fdr', 'fip37_frac_mod', 'fio1_frac_mod', 'col0_frac_mod',
    'g_stat', 'hom_g_stat', 'fip37_vs_fio1_g_stat', 'fip37_vs_col0_g_stat',
    'fio1_vs_col0_g_stat', 'fip37_vs_fio1_fdr', 'fip37_vs_col0_fdr',
    'fio1_vs_col0_fdr', 'unmod_mean', 'unmod_std', 'mod_mean', 'mod_std',
    'ks_stat'
]
```

```python
tmp_bed = os.path.join(os.environ['TMPDIR'], 'yanocomp_miclip_dist.bed')

pybt.BedTool(snakemake.input.yanocomp_sites[0]).closest(
    b=snakemake.input.miclip_peaks,
    s=True, D='a'
).saveas(tmp_bed)


MICLIP_COLUMNS = ['miclip_chrom', 'miclip_start', 'miclip_end', 'miclip_name', 'miclip_score', 'miclip_strand', 'miclip_dist']

fip37_fio1_res_miclip_dist = pd.read_csv(
    tmp_bed, sep='[\t:\[\],]+',
    names=YANOCOMP_COLUMNS + MICLIP_COLUMNS, engine='python'
).drop_duplicates(['chrom', 'start', 'end', 'gene_id', 'kmer', 'strand'])
```

```python
COMPS = ['fip37_vs_fio1', 'fip37_vs_col0', 'fio1_vs_col0']

for comp in COMPS:
    comp_sig = fip37_fio1_res_miclip_dist.query(f'{comp}_fdr < 0.05').iloc[:, [0, 1, 2, 3, 5, 6]]
    comp_sig.to_csv(f'yanocomp/{comp}.posthoc.bed', sep='\t', header=False, index=False)

for comp1, comp2 in it.permutations(COMPS, r=2):
    # for difference sets only use examples with miCLIP support
    comp_sig = fip37_fio1_res_miclip_dist.query(f'{comp1}_fdr < 0.05 & {comp2}_fdr > 0.05 & abs(miclip_dist) < 5').iloc[:, [0, 1, 2, 3, 5, 6]]
    comp_sig.to_csv(f'yanocomp/{comp1}__not__{comp2}.posthoc.bed', sep='\t', header=False, index=False)
```

```python
fip37_fio1_res_miclip_dist['FIP37_dependent'] = fip37_fio1_res_miclip_dist['fip37_vs_col0_fdr'] < 0.05
fip37_fio1_res_miclip_dist['FIO1_dependent'] = fip37_fio1_res_miclip_dist['fio1_vs_col0_fdr'] < 0.05
fip37_fio1_res_miclip_dist['miCLIP'] = fip37_fio1_res_miclip_dist['miclip_dist'].abs() < 5

display_formatted_markdown(
    MD_TEXT[0],
    n_sites_total=len(fip37_fio1_res_miclip_dist),
    perc_fip=fip37_fio1_res_miclip_dist.FIP37_dependent.mean() * 100,
    num_fip=fip37_fio1_res_miclip_dist.FIP37_dependent.sum(),
    perc_fio=fip37_fio1_res_miclip_dist.FIO1_dependent.mean() * 100,
    num_fio=fip37_fio1_res_miclip_dist.FIO1_dependent.sum(),
)

display_formatted_markdown(
    MD_TEXT[1],
    fip_perc_miclip=fip37_fio1_res_miclip_dist.query('FIP37_dependent').miCLIP.mean() * 100,
    fip_num_miclip=fip37_fio1_res_miclip_dist.query('FIP37_dependent').miCLIP.sum(),
    num_fio_only=len(fip37_fio1_res_miclip_dist.query('FIO1_dependent & ~FIP37_dependent')),
    fio_only_perc_miclip=fip37_fio1_res_miclip_dist.query('FIO1_dependent & ~FIP37_dependent').miCLIP.mean() * 100,
    fio_only_num_miclip=fip37_fio1_res_miclip_dist.query('FIO1_dependent & ~FIP37_dependent').miCLIP.sum(),
)

upset_data = fip37_fio1_res_miclip_dist[['FIP37_dependent', 'FIO1_dependent', 'miCLIP']]
upset_data = upset_data.groupby(['FIP37_dependent', 'FIO1_dependent']).apply(
    lambda g: pd.Series((len(g), sum(g.miCLIP), g.miCLIP.mean() * 100), index=['total', 'miclip', 'miclip_perc']))
upset = plot_upset(
    upset_data.total, facecolor='#0072b2', sort_by='degree', sort_categories_by=None, element_size=100,
    intersection_plot_elements=4, totals_plot_elements=1,
)
upset['intersections'].bar(x=np.arange(len(upset_data)), height=upset_data.miclip.values[[0, 2, 1, 3]], width=0.5, color=pal[1], zorder=10)
upset['intersections'].legend(['False', 'True'], title='miCLIP support')
i = 0
for h, _, p in upset_data.iloc[[0, 2, 1, 3]].itertuples(index=False):
    upset['intersections'].annotate(s=f'{h:.0f}\n({p:.1f}%)', xy=(i, h), ha='center', va='bottom')
    i += 1
plt.savefig(snakemake.output.site_upset)
plt.show()
```

```python
m = fip37_fio1_res_miclip_dist.query('FIO1_dependent')
m = m.assign(
    fip_lod=np.abs(np.log(m.fip37_frac_mod + 1e-3) - np.log(m.col0_frac_mod + 1e-3)),
    fio_lod=np.abs(np.log(m.fio1_frac_mod + 1e-3) - np.log(m.col0_frac_mod + 1e-3)),
)

display_formatted_markdown(
    MD_TEXT[2],
    fio_perc_fip=fip37_fio1_res_miclip_dist.query('FIO1_dependent').FIP37_dependent.mean() * 100,
    fio_num_fip=fip37_fio1_res_miclip_dist.query('FIO1_dependent').FIP37_dependent.sum(),
)


m = pd.melt(
    m, id_vars=[], value_vars=['fip_lod', 'fio_lod'],
    var_name='Genotype',
    value_name='Effect size (absolute log ratio)'
)

fig, ax = plt.subplots(figsize=(5, 5))
sns.boxplot(
    x='Genotype',
    y='Effect size (absolute log ratio)',
    data=m, palette=[pal[2], pal[1]], showfliers=False
)
ax.set_xticklabels(['fip37-4', 'fio1-1'])
plt.savefig(snakemake.output.site_effect_size)
plt.show()
```
